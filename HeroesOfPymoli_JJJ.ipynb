## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spender = purchase_data
top_spender["Price_sum"] = top_spender["Price"]
top_spender["Count"] = 0
top_spender = top_spender.groupby(["SN"]).agg({"Price":"mean", "Price_sum": "sum", "Count":"size"})

top_spender = top_spender.sort_values(by=["Price_sum"], ascending=False)

top_spender = top_spender.rename(columns={"Count": "Purchase Count", "Price": "Average Purchase Price", 
                                          "Price_sum": "Total Purchase Value"}) 

top_spender_summary = pd.DataFrame(top_spender)
top_spender_summary =top_spender_summary[["Purchase Count", "Average Purchase Price"
                            ,"Total Purchase Value"]]

top_spender_summary["Average Purchase Price"] =top_spender_summary["Average Purchase Price"].map('${:,.2f}'.format)
top_spender_summary["Total Purchase Value"] =top_spender_summary["Total Purchase Value"].map('${:,.2f}'.format)
top_spender_summary.head(10)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_popular_items = purchase_data
most_popular_items ["Price_sum"] =most_popular_items["Price"]
most_popular_items ["Count"] = 0
most_popular_items =most_popular_items.groupby(["Item ID"]).agg({"Item Name":"min", "Price":"min"
                                                                 ,"Price_sum":"sum", "Count":"size"})

most_popular_items =most_popular_items.sort_values(by=["Count"], ascending=False)

most_popular_items =most_popular_items.rename(columns={"Count": "Purchase Count"
                                                      ,"Price" : "Item Price"
                                                      ,"Price_sum":"Total Purchase Value"
                                                      })

most_popular_summary = pd.DataFrame(most_popular_items)

most_popular_summary =most_popular_summary[["Item Name","Purchase Count", "Item Price","Total Purchase Value"]]
most_popular_summary["Item Price"] =most_popular_summary["Item Price"].map('${:,.2f}'.format)
most_popular_summary["Total Purchase Value"] =most_popular_summary["Total Purchase Value"].map('${:,.2f}'.format)

most_popular_summary.head(10)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profitable_items = purchase_data
most_profitable_items["Price_sum"] = most_profitable_items["Price"]
most_profitable_items["Count"] = 0
most_profitable_items = most_profitable_items.groupby(["Item ID"]).agg({"Item Name":"min","Price":"min"
                                                                        ,"Price_sum":"sum","Count":"size"})
                                                                                  
most_profitable_items = most_profitable_items.sort_values(by=["Price_sum"], ascending=False)

most_profitable_items = most_profitable_items.rename(columns={"Count": "Purchase Count"
                                                      ,"Price" : "Item Price"
                                                      ,"Price_sum":"Total Purchase Value"
                                                      })

most_profitable_summary = pd.DataFrame(most_profitable_items)

most_profitable_summary = most_profitable_summary[["Item Name"
                                        ,"Purchase Count"
                                        ,"Item Price"
                                        ,"Total Purchase Value"                       
                                     ]]

most_profitable_summary["Item Price"] = most_profitable_summary["Item Price"].map('${:,.2f}'.format)
most_profitable_summary["Total Purchase Value"] = most_profitable_summary["Total Purchase Value"].map('${:,.2f}'.format)
most_profitable_summary.head(10)
```
